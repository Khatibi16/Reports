### **1. Executive Summary**

*   **Investment Recommendation:** **Hold**. Expedia Group is demonstrating operational improvements and strong growth in its B2B segment. However, the stock appears to have priced in much of the near-term upside. Intense competition and macroeconomic sensitivity warrant a neutral stance pending clearer signals of sustained margin expansion and successful integration of new strategic initiatives.
*   **Case Framing:**
    *   **Base Case:** The company achieves high single-digit revenue growth driven by its B2B segment and maintains stable margins. Valuation remains range-bound as competitive pressures from Booking Holdings and Airbnb cap multiple expansion.
    *   **Bull Case:** Successful AI integration, loyalty program adoption, and growth in high-margin "Tours and Activities" drive market share gains and meaningful margin improvement. The B2B segment continues its high-growth trajectory, becoming a larger, more profitable portion of the business.
    *   **Bear Case:** A cyclical downturn in travel spending, increased competition leading to price wars, or execution missteps with recent restructuring and tech platform integration cause growth to stagnate and margins to compress.
*   **What Must Go Right:** Management must sustain its recent operational discipline, successfully execute its technology platform consolidation, continue scaling the high-margin B2B business, and fend off intense competition without sacrificing profitability.
*   **What Would Break the Thesis:** A material slowdown in the B2B growth engine, evidence of market share loss in core B2C segments, or a failure to translate recent operational restructuring into sustained free cash flow growth would invalidate the hold thesis.
*   **Confidence Level:** Medium. The company's financial trajectory is positive, but the industry is volatile and highly competitive.

### **2. Company & Business Overview**

Expedia Group, Inc. is a global online travel company (OTC) providing a full range of travel and advertising services. The business model is centered on earning commissions, transaction fees, and advertising revenue from facilitating travel bookings.

The company operates through three primary segments as detailed in its financial reports:
*   **B2C (Business-to-Consumer):** The largest segment, encompassing brands like Expedia.com, Hotels.com, and Vrbo. It serves customers directly, offering lodging, air travel, car rentals, and activities. This segment is driven by marketing efficiency and brand recognition.
*   **B2B (Business-to-Business):** This fast-growing segment provides travel technology and supply to other businesses, including airlines, travel agencies, and financial institutions. It leverages Expedia's vast inventory and technological infrastructure, generating higher margins than the B2C segment. Recent performance shows this is a key growth driver.
*   **trivago:** A majority-owned hotel metasearch engine that generates advertising revenue from referrals to OTCs and travel service providers. This segment has faced headwinds and declining revenue, as shown in the `EXPE.O | Statement App | Income Statement` document.

The business is structurally tied to the global travel industry, making it sensitive to macroeconomic cycles, geopolitical stability, and consumer discretionary spending.

### **3. Financial Performance & Trends**

*Data sourced from the provided `Company Fundamentals - Financial Summary` for FY2022-2024.*

*   **Revenue Growth:** Revenue grew from $11.7B in 2022 to a projected $13.7B in 2024, a CAGR of 8.2%. This reflects a normalization of growth following the post-pandemic travel surge.
*   **EBITDA Growth:** EBITDA is projected to grow from $2.0B in 2022 to $2.5B in 2024, a CAGR of 11.9%, indicating improving operational leverage and profitability. The EBITDA margin is forecast to expand from 17.0% to 18.1% over the same period.
*   **EPS Growth:** Diluted EPS shows significant growth, from $2.18 in 2022 to a projected $8.95 in 2024. This substantial increase is driven by higher net income and a reduction in shares outstanding from aggressive buybacks.
*   **Margin Evolution:** Gross profit margin is robust and expanding, from 79.0% in 2022 to a projected 83.3% in 2024. The operating margin is also showing steady improvement from 10.2% to 12.0%.
*   **Return Metrics:** Return on Invested Capital (ROIC) shows a strong positive trend, increasing from 4.8% in 2022 to a projected 15.8% in 2024. Return on Equity (ROE) is exceptionally high, projected at 79.8% for 2024, though this is flattered by a relatively small equity base due to high leverage and historical buybacks.
*   **Balance Sheet Strength:** The balance sheet carries significant leverage. The Total Debt to Total Equity ratio is projected to be 223.9% for 2024. While the company holds a substantial cash balance ($4.2B in 2023), the high leverage remains a key risk factor.

### **10. Competitive Landscape**

*   **Industry Structure:** The online travel industry is a near-duopoly dominated by Expedia Group and Booking Holdings (BKNG.OQ), with Airbnb (ABNB.OQ) being a major force in alternative accommodations. Competition is intense and primarily fought through marketing spend, technology, and supply acquisition.
*   **Key Competitors:**
    *   **Booking Holdings:** Expedia's primary direct competitor, with a stronger international presence and historically higher margins.
    *   **Airbnb:** Dominates the alternative accommodations space and is increasingly competing in hotels and experiences.
    *   **Google:** A major source of traffic but also a competitive threat with its own travel search products.
    *   **Direct Suppliers:** Hotels and airlines are continually trying to encourage direct bookings to avoid paying commissions to OTCs.
*   **Competitive Advantages:** Expedia's key advantages are its vast network of travel supply, established brand portfolio, and significant technology infrastructure. Its growing B2B business creates a stickier revenue stream that is less susceptible to consumer marketing wars.
*   **Pricing Power:** Pricing power is limited due to the high transparency and competition in the market. Value is created through scale, efficiency, and packaging of services.

### **12. Valuation**

Based on the comparable analysis, Expedia appears reasonably valued relative to its peers. The 9.2x forward EV/EBITDA multiple is not demanding for a company with projected double-digit EBITDA growth and strong FCF generation.

*   **Upside Scenario:** If the B2B segment continues its >20% growth and margins expand, the market could re-rate the stock closer to a 11-12x multiple, suggesting a price target of ~$280-300, representing a 20-30% upside.
*   **Downside Scenario:** A recessionary environment could see multiples compress to 7-8x, suggesting a price target of ~$185-210, representing a 10-20% downside.

The current valuation seems to price in the base case of steady execution with moderate growth.

### **13. Risks & Disconfirming Evidence**

*   **Fundamental Risks:** High sensitivity to the macroeconomic cycle and consumer discretionary spending. A recession would materially impact bookings and revenue.
*   **Competitive Risks:** Intense and sustained competition from Booking Holdings, Google, and Airbnb could erode market share or force higher marketing spending, compressing margins.
*   **Execution Risks:** The company is undergoing restructuring, including layoffs and leadership changes. Failure to execute cleanly could disrupt operations and morale.
*   **Financial Risks:** The balance sheet has high total leverage (Total Debt/Equity > 200%). While manageable with current cash flows, an economic downturn could make this leverage problematic.
*   **Signals that would invalidate the thesis:**
    *   Quarterly reports showing a deceleration in B2B revenue growth below 15%.
    *   Loss of market share in lodging gross bookings to competitors.
    *   A decline in FCF margins due to increased marketing spend without a corresponding increase in revenue.

### **14. Final Investment View**

*   **What is priced in:** The market has priced in the post-pandemic travel recovery, mid-to-high single-digit growth, and the benefits of recent operational improvements. The current valuation reflects Expedia's position as a major player but not a best-in-class operator compared to Booking Holdings.
*   **What is not priced in:** The full long-term potential of the B2B segment to transform the business mix toward higher, more stable margins is likely underappreciated. Conversely, the market may also be underestimating the long-term competitive threat from AI-native travel startups and Google's increasing encroachment.
*   **Risk-Reward Skew:** The risk-reward appears balanced at the current price. The potential for a re-rating based on B2B growth and margin expansion (bull case) is counterweighed by significant cyclical and competitive risks (bear case).
*   **Final Recommendation:** **Hold**. We will continue to monitor the key drivers, particularly B2B segment performance and margin trajectory. We would become more constructive on a material pullback in the share price or clear evidence that the company is taking sustainable market share. A change to a Sell recommendation would be warranted if we see margin erosion and a sustained deceleration in growth.
