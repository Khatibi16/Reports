**INVESTMENT MEMORANDUM: Euronext NV (ENX.PA)**<br>
**Date:** February 10, 2026<br>
**Subject:** Investment Analysis of Euronext NV<br>

### 1. Executive Summary

*   **Investment Recommendation: Buy.** Euronext NV stands out as an attractive investment opportunity, combining defensive business characteristics with clear catalysts for growth and shareholder returns. The company has delivered its sixth consecutive quarter of double-digit growth, driven by a successful strategic pivot to non-volume-related businesses (now 60% of revenue) and disciplined cost control. The balance sheet is strong, with leverage comfortably within the target range (1.5x Net Debt to EBITDA), enabling a proactive capital allocation policy. This includes a stable dividend, a new €250 million share buyback program, and a disciplined, EPS-accretive acquisition of ATHEX Group. At the current valuation, we believe the market underappreciates the durability of its diversified revenue streams, the impact of its enhanced shareholder return policy, and its potential for continued value-accretive consolidation.

*   **Case Framing & Price Targets:** The current stock price is **€120.40** (as of 10-Feb-2026). Our price targets are based on the company's continued execution of its "Innovate for Growth 2027" strategy.
    *   **Base Case (60% Probability) | Price Target: €135:** Euronext successfully closes and integrates the ATHEX Group acquisition, delivering on guided synergies. The company executes its share buyback program and continues to grow its non-volume-related businesses, leading to a valuation in line with its high-quality FMI peers.
    *   **Bull Case (25% Probability) | Price Target: €155:** Growth in the high-margin Capital Markets and Data Solutions segment outpaces expectations. The company leverages its strong cash flow and balance sheet for further value-accretive bolt-on M&A or enhanced shareholder returns, leading to a positive re-rating of its valuation multiple.
    *   **Bear Case (15% Probability) | Price Target: €105:** A sharp and prolonged macroeconomic downturn significantly suppresses trading volumes and listing activity. The ATHEX acquisition fails to deliver synergies, or cost discipline slips, leading to margin compression and a pause in capital return programs.

*   **What Must Go Right:** Management must maintain its demonstrated cost discipline, successfully integrate the ATHEX Group to realize guided synergies, and continue growing its recurring revenue businesses in data and post-trade services. The execution of the announced share buyback program is key to the shareholder return thesis.

*   **What Would Break the Thesis:** A significant reversal in the growth of non-volume-related revenues, a failure to control costs, or a major strategic misstep in M&A that materially increases leverage outside the target range would invalidate the investment thesis.

*   **Confidence Level: Medium-High.** Confidence is high given the strong recent performance, clear strategic plan, manageable leverage, and shareholder-friendly capital allocation. The bolt-on nature of the current M&A presents lower risk than a large-scale transformative deal.

### 2. Company & Business Overview

Euronext NV is the leading pan-European market infrastructure provider, operating a "federal model" that connects local economies to global capital markets. Its primary exchanges are located in Amsterdam, Brussels, Dublin, Lisbon, Milan, Oslo, and Paris. The company's business is diversified across several interconnected segments:
*   **Listing:** The core traditional business of providing a venue for companies to list their securities and raise capital.
*   **Trading (Equity, FICC, Derivatives):** Operates cash and derivatives markets across multiple asset classes. A significant portion of this revenue is volume-dependent.
*   **Post-Trade (Clearing & Settlement):** Provides clearing services through Euronext Clearing and settlement/custody services. These activities generate stable and recurring fee income.
*   **Capital Markets and Data Solutions:** A key growth engine, this segment includes market data sales, advanced analytics, and technology services. These revenues are typically high-margin and subscription-based.

This diversified model, with a growing share of non-volume-related revenue (60% in Q3 2025), provides resilience across different market cycles.

**Latest Quarter Results (Q3 2025, Reported EUR)**
The Q3 2025 results marked the sixth consecutive quarter of double-digit growth, underscoring the success of the company's strategic initiatives.
*   **Revenue and Income:** €438.1 million, up 10.6% YoY, driven by strong performance in non-volume businesses like Capital Markets and Data Solutions (+13.9%) and resilient trading revenues.
*   **Cost Discipline:** Underlying operating costs (excluding D&A) were well-managed, leading the company to upgrade its full-year 2025 cost guidance to €660 million (from €670 million).
*   **Profitability:** Adjusted EBITDA grew 12.6% to €276.7 million, with the adjusted EBITDA margin expanding by 1.2 percentage points to 63.2%.
*   **Adjusted EPS:** Grew to €1.68, reflecting strong operational performance and accretion from prior M&A.

### 4. Free Cash Flow & Capital Efficiency

Euronext's business model is highly cash-generative, providing significant financial flexibility.
*   **Cash Flow Generation:** The company generated a strong Net Operating Cash Flow of €401.0 million in Q3 2025. Excluding certain working capital movements, the cash conversion rate from EBITDA was 99.9%, demonstrating exceptional efficiency.
*   **Balance Sheet Strength:** As of September 30, 2025, the Net Debt to EBITDA ratio stood at **1.5x**, which is comfortably within the company's target range of 1.0x to 2.0x. This strong balance sheet enables the company to pursue its strategic objectives without taking on undue financial risk.

### 5. Capital Allocation Analysis (Acquisition, Dividend, Buyback)

Euronext has a clear and proactive capital allocation strategy ("Innovate for Growth 2027") that balances investment for growth with direct returns to shareholders.

1.  **Disciplined M&A (Acquisition):** Euronext is pursuing a **voluntary share exchange offer for ATHEX Group** (the Athens Stock Exchange). This is a strategically compelling, bolt-on acquisition that boosts the attractiveness of the Greek market, is supported by the ATHEX board, and is expected to be **EPS accretive in year one**. The deal targets €12 million in annual cash synergies by year three. This action aligns with the strategy of pursuing value-creative M&A (ROCE > WACC).
2.  **Ordinary Returns (Dividend):** The company maintains a clear policy of returning capital to shareholders through a **dividend payout of 50% of reported earnings**. This provides a stable and predictable income component to the investment case.
3.  **Special Returns (Buyback):** Enabled by its strong cash generation and fast deleveraging, Euronext has launched a **share repurchase program for a maximum of €250 million**, starting in November 2025. This represents approximately 2% of the outstanding shares and demonstrates management's confidence in the company's prospects and commitment to enhancing shareholder value.

### 6. Ownership & Governance

*   **Ownership Structure:** Euronext has a unique "federal" governance model with a group of core strategic shareholders including Cassa Depositi e Prestiti (Italy) and Caisse des Dépôts et Consignations (France), each holding around 8%. This structure reflects its pan-European nature and provides a stable long-term shareholder base. The majority of the company is owned by a diverse group of international institutional investors.
*   **Governance:** The board and management are experienced in navigating the complex European financial landscape. The primary governance focus for investors should be on the disciplined execution of the capital allocation strategy.

### 7. Insider Activity

The provided insider report lists several individuals with very small holdings. One insider, Daryl Byrne, shows a minor reduction in his position (-400 shares), which is not material. Other insiders show small new positions. There is no evidence of widespread or significant insider selling that would indicate a lack of confidence in the company's prospects.

### 8. Litigation, Regulatory & Legal Risk

As a central market infrastructure provider, Euronext is subject to extensive regulation by national and European authorities. The primary regulatory risk in the near term is securing the necessary approvals for the ATHEX Group acquisition. The provided documents do not indicate any material ongoing litigation.

This analysis is based on the company's detailed Q3 2025 results presentation, which provides a comprehensive overview of recent performance and strategic initiatives. It is supplemented by Refinitiv financial data, ownership summaries, and peer comparisons.

### 10. Competitive Landscape

Euronext's main competitors are other major exchange groups, including **Deutsche Boerse**, **London Stock Exchange Group**, **Nasdaq**, and **Intercontinental Exchange (ICE)**. Competition is centered on attracting listings, providing liquid and efficient trading venues, and offering value-added data and technology services.

### 11. Comparable Company Analysis

Based on the provided factsheet, Euronext trades at a **P/E (TTM) of approximately 19.2x**. Its Net Debt/EBITDA ratio of **1.5x** (per the Q3 presentation) is healthy and compares favorably to many of its peers, providing greater financial flexibility. A full peer valuation table was not provided, but the current multiple appears reasonable for a high-quality FMI business with a strong growth and capital return profile.

### 12. Valuation

The price targets reflect our view that Euronext's consistent execution and shareholder-friendly policies will drive a re-rating of the stock.
*   **Valuation Approach:** The price targets are based on a forward P/E multiple applied to our earnings projections, which are informed by company guidance and recent performance.
    *   **Base Case (€135):** Assumes a forward P/E of ~19-20x, in line with its current valuation, as the market digests the ATHEX acquisition and buyback program.
    *   **Bull Case (€155):** Assumes a forward P/E re-rating to ~21-22x, driven by above-plan synergy capture and sustained double-digit growth in recurring revenues.
    *   **Bear Case (€105):** Assumes a multiple contraction to ~16-17x due to a cyclical downturn or a failure to execute on its strategic initiatives.

### 13. Risks & Disconfirming Evidence

*   **Macroeconomic Sensitivity:** A significant economic downturn could reduce trading volumes, corporate listing activity, and demand for data services, impacting revenue growth.
*   **M&A Integration Risk:** While the ATHEX deal is a bolt-on, all acquisitions carry integration risks, including the risk of not achieving projected synergies.
*   **Cost Control:** The company's strong profitability relies on continued cost discipline. Any unexpected rise in the cost base would pressure margins.
*   **Competitive Threats:** The European exchange landscape is highly competitive. Increased competition could pressure trading fees or require higher investment to maintain market share.

### 14. Final Investment View

*   **What is priced in:** The market currently prices Euronext as a stable, well-managed European exchange operator. The current valuation reflects its solid, but not spectacular, historical performance.
*   **What is not priced in:** The market appears to be underappreciating the sustainability of the recent double-digit growth, the positive impact of the shift to 60% non-volume-related revenues, and the accretive power of a combined disciplined M&A and a significant share buyback program.
*   **Risk-Reward Skew:** The risk-reward profile is favorably skewed to the upside. The downside is cushioned by the company's resilient business model and strong balance sheet, while the upside is driven by clear, tangible catalysts in the form of accretive M&A and shareholder returns.
*   **Final Recommendation: Buy.** Euronext presents a compelling investment case based on a proven strategy, strong execution, a healthy balance sheet, and a clear commitment to enhancing shareholder value. We recommend purchasing the shares at the current price.
